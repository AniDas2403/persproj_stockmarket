## Normalizing Datasets

```python
x1min, x2min, x3min, x4min, x5min, x6min, x7min, x8min, x9min, x10min, x11min, x12min, x13min = np.amin(data['Open']), np.amin(data['High']), np.amin(data['Low']), np.amin(data['high_low']), np.amin(data['open_close']), np.amin(data['mvavgo7']), np.amin(data['mvavgc7']), np.amin(data['mvavgo14']), np.amin(data['mvavgc14']), np.amin(data['mvavgo21']), np.amin(data['mvavgc21']), np.amin(data['sd7o']), np.amin(data['sd7c'])
x1max, x2max, x3max, x4max, x5max, x6max, x7max, x8max, x9max, x10max, x11max, x12max, x13max = np.amax(data['Open']), np.amax(data['High']), np.amax(data['Low']), np.amax(data['high_low']), np.amax(data['open_close']), np.amax(data['mvavgo7']), np.amax(data['mvavgc7']), np.amax(data['mvavgo14']), np.amax(data['mvavgc14']), np.amax(data['mvavgo21']), np.amax(data['mvavgc21']), np.amax(data['sd7o']), np.amax(data['sd7c'])
d1, d2, d3, d4, d5, d6, d7, d8, d9, d10, d11, d12, d13 = x1max-x1min, x2max-x2min, x3max-x3min, x4max-x4min, x5max-x5min, x6max-x6min, x7max-x7min, x8max-x8min, x9max-x9min, x10max-x10min, x11max-x11min, x12max-x12min, x13max-x13min
```

```python
def normalize_x(X1, X2, X3, X4, X5, X6, X7, X8, X9, X10, X11, X12, X13):
    Z = []
    for x in range(len(X1)):
        q1 = (X1[x] - x1min)/d1
        q2 = (X2[x] - x2min)/d2
        q3 = (X3[x] - x3min)/d3
        q4 = (X4[x] - x4min)/d4
        q5 = (X5[x] - x5min)/d5
        q6 = (X6[x] - x6min)/d6
        q7 = (X7[x] - x7min)/d7
        q8 = (X8[x] - x8min)/d8
        q9 = (X9[x] - x9min)/d9
        q10 = (X10[x] - x10min)/d10
        q11 = (X11[x] - x11min)/d11
        q12 = (X12[x] - x12min)/d12
        q13 = (X13[x] - x13min)/d13
        Z.append([q1,q2,q3,q4,q5,q6,q7,q8,q9,q10,q11,q12,q13])
    Z = np.asarray(Z)
    return Z
```

```python
Ymax = np.amax(data['Close'])
Ymin = np.amin(data['Close'])
D = Ymax - Ymin
```

```python
def normalize_y(Y):
    Z = []
    for x in range(len(Y)):
        Z.append((Y[x] - Ymin)/D)                                     
    Z = np.asarray(Z)
    return Z
```
